```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
df.head()
```

```python
y=df.Survived
df=df.drop(['Survived'],axis=1)
df=df.drop(['Name'],axis=1)
df = df.drop(['PassengerId'],axis=1)
df_test = df_test.drop(['Name'],axis=1)
p_id = df_test.PassengerId
df_test = df_test.drop(['PassengerId'],axis=1)
sex_det= df_test.Sex
df.head()
```

```python
df = pd.get_dummies(df)
df_test= pd.get_dummies(df_test)
df.head()
```

```python
df_test.head()
```

```python
final_train, final_test = df.align(df_test,join='left',axis=1)
```

```python
final_train.head()
```

```python
final_test.head()
```

```python
from sklearn.impute import SimpleImputer
df_imputer= SimpleImputer(missing_values=np.nan, strategy='mean', fill_value=0)
```

```python
df_imputed = pd.DataFrame(df_imputer.fit_transform(final_train))
df_imputed.describe()
```

```python
model = XGBRegressor(n_estimators=5000,learning_rate=0.05,objective ='reg:logistic')
df_imputed.head()
```

```python
model.fit(df_imputed,y,verbose=False)
```

```python
df_test_imputer = SimpleImputer(missing_values=np.nan, strategy='mean', fill_value=0)
```

```python
final_test.head()
```

```python
df_test_new = pd.DataFrame(df_imputer.transform(final_test))
```

```python
df_test_new.head()
```

```python
predictions = model.predict(df_test_new)
```

```python
answer=pd.DataFrame()
```

```python
answer['PassengerId']=p_id
answer['Survived']=np.round(predictions).astype('int')
```

```python
answer.head()
```

```python
answer.to_csv('submission.csv', index=False)
```
